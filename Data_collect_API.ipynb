```python
import pandas as pd
import numpy as np
import requests
import json
import time
from tqdm.auto import tqdm as tq
```

```python
coins = requests.get('https://api.coingecko.com/api/v3/coins/list?include_platform=true')
```

```python
y = json.loads(coins.text)
print(len(y))
coinid = [coin["id"] for coin in y]
print(coinid)
```

```python
coinid1 = coinid[1000:2000]
```

```python
x1 = requests.get('https://api.coingecko.com/api/v3/coins/bitcoin?localization=false&tickers=true&market_data=true&community_data=true&developer_data=true&sparkline=true')
df1 = pd.DataFrame.from_dict([json.loads(x1.text)])
df1.head()
```

```python
for id, pos in zip(tq(coinid1),range(len(coinid1))):
  time.sleep(2)
  res = requests.get("https://api.coingecko.com/api/v3/coins/"+id+"?localization=false&tickers=true&market_data=true&community_data=true&developer_data=true&sparkline=true")
  df1 = pd.concat([df1, pd.DataFrame.from_dict([json.loads(res.text)])], ignore_index=True)
  df1.reset_index()
```

```python
df1.to_excel("data2.xlsx")
```
